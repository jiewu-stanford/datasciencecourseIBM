```python
import pandas as pd
# hence I have to import the data pre-compiled CSV file
!wget -O Geospatial_Coordinates.csv http://cocl.us/Geospatial_data/Geospatial_Coordinates.csv
```

```python
# read in the geo-coordinate table
df_latlng = pd.read_csv('Geospatial_Coordinates.csv')
df_latlng.columns = ['PostalCode', 'Borough', 'Neighbourhood']
df_latlng.head()
```

```python
# run the web table scraping notebook and read in the postal code table
%run ./WikiTableScrape.ipynb
df_pc = pd.read_csv('PCodeTable.csv')
df_pc.head()
```

```python
# merge the two tables
df = pd.merge(df_pc, df_latlng, on='PostalCode', how = 'inner')
df.columns = ['PostalCode', 'Borough', 'Neighbourhood', 'Latitude', 'Longitude']
df.head()
```

```python
df.to_csv('GeoPCodeTable.csv', index=False)
```
